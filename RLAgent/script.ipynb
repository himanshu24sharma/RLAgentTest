```python

with WarehouseEnv(base_url="http://localhost:8000").sync() as client:

    client.reset()

    # Move right
    obs = client.step(WarehouseAction(action=3))
    print(f"Move right → pos=({obs.observation.agent_x},{obs.observation.agent_y}) reward={obs.observation.reward:.2f}")

    # Invalid pick (nothing here)
    obs = client.step(WarehouseAction(action=4))
    print(f"Invalid pick → reward={obs.observation.reward:.2f} (should be -0.21)")

    # Invalid deliver (not holding)
    obs = client.step(WarehouseAction(action=5))
    print(f"Invalid deliver → reward={obs.observation.reward:.2f} (should be -0.21)")
```

```python
with WarehouseEnv(base_url="http://localhost:8000").sync() as client:
    from baseline import greedy_agent

    obs = client.reset()
    print("\n=== Baseline agent via HTTP ===")

    while not obs.observation.done:
        action = greedy_agent(obs.observation.model_dump())
        obs = client.step(WarehouseAction(action=action))

    print(f"Items delivered : {obs.observation.items_delivered}/{obs.observation.total_items}")
    print(f"Steps used      : {obs.observation.steps_elapsed}/{obs.observation.max_steps}")
    print(f"Completed       : {obs.observation.done}")
```
